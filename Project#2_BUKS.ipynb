```python
k1m = 0.001    #variant
#k1m = 0.005
#k1m = 0.01
#k1m = 0.015
#k1m = 0.02
k3 = 0.0032
k2 = 2
k3m= 0.002
x = [round((i+0.001)/1000,3) for i in range(384)]
y = [k3/k3m*x[i] for i in range(384)]
z = [1-x[i]-y[i] for i in range(384)]
k1 = [(k1m*x[i]+k2*x[i]*(z[i])**2)/z[i] for i in range(384)]
a11 = [-k1[i]-k1m-k3-k2+4*k2*x[i]-3*k2*(x[i])**2+2*k2*y[i]-4*k2*y[i]*x[i]-k2*(y[i])**2 for i in range(384)]
a12 = [-k1[i]+k3m+2*k2*x[i]-2*k2*(x[i])**2-2*k2*x[i]*y[i] for i in range(384)]
a21 = k3 
a22 = -k3m
sp = [a11[i]+a22 for i in range(384)]
determ = [a11[i]*a22-a12[i]*a21 for i in range(384)]
index = 1
yh = []
xh = []
k1h = []
ysn = []
xsn = []
k1sn = []
while index < 384:
    if sp[index]*sp[index-1] <= 0:
        yh.append(y[index])
        xh.append(x[index])
        k1h.append(k1[index])
    if determ[index]*determ[index-1] <= 0:
        ysn.append(y[index])
        xsn.append(x[index])
        k1sn.append(k1[index])
    index += 1
print(len(xh))
print(len(xsn))
print("hopf")
print(yh)
print(xh)
print(k1h)
mh = len(xh)
msn = len(xsn)
print("sn")
print(ysn)
print(xsn)
print(k1sn)
fig = plt.figure()
plt.plot(k1, x, label = u'x')
plt.plot(k1, y, label = u'y', linestyle='--')
plt.plot(k1h,yh, label = u'hopf(y(k1))', linestyle = 'None', marker='*', markersize = '15')
plt.plot(k1sn,ysn, label = u'saddle-node(y(k1))', linestyle = 'None', marker='d')
plt.plot(k1h,xh, label = u'hopf(x(k1))', linestyle = 'None', marker='*', markersize = '15', color='b')
plt.plot(k1sn,xsn, label = u'saddle-node(x(k1))', linestyle = 'None', marker='d',color='m')
plt.title('Dependence on Parameter k1, k1m = 0.001',fontsize=16)
plt.ylabel('x,y',fontsize=14)
plt.xlabel('k1',fontsize=14)
x1,x2,y1,y2 = plt.axis()
plt.axis((x1,x2,0,1))
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.integrate import odeint

def model(z,t,k1,k1m,k2,k3,k3m):
    x = z[0]
    y = z[1]
    dxdt = k1*(1-x-y)-k1m*x-k3*x+k3m*y-k2*((1-x-y)**2)*x
    dydt = k3*x-k3m*y
    dzdt = [dxdt,dydt]
    return dzdt

k1m = -0.5 # I (-0.5,0.08) image below  
# sn [0.103; 0.1648]
k1 = 0.08049722622234359 
k3 = 0.0032
k2 = 2
k3m= 0.002
#z = [0.0950,0.1520]
#z = [0,0]
z = [0.15, 0.2]
t = np.linspace(0,100,1000000)   
# max step 1000000
res = odeint(model, z, t, args = (k1,k1m,k2,k3,k3m))
xxx = []
yyy = []
for i in range(len(res)):
    xxx.append(res[i][0])
    yyy.append(res[i][1])
    
fig1 = plt.figure()
plt.plot(xxx, yyy, label = u'y(x)')
plt.title('y(x)',fontsize=16)
plt.ylabel('y',fontsize=14)
plt.xlabel('x',fontsize=14)
x1,x2,y1,y2 = plt.axis()
plt.axis((x1,x2,y1,y2))
plt.legend()
plt.grid(True)
plt.show()

fig = plt.figure()
plt.plot(t, xxx, label = u'x(t)')
plt.plot(t, yyy, label = u'y(t)', linestyle='--')
plt.title('Dependence on t',fontsize=16)
plt.ylabel('x, y',fontsize=14)
plt.xlabel('t',fontsize=14)
x1,x2,y1,y2 = plt.axis()
plt.axis((x1,x2,y1,y2))
plt.legend()
plt.grid(True)
plt.show()
```
