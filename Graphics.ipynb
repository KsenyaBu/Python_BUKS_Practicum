```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.legend_handler import HandlerLine2D
#from pylab import *

verlet = [0.0566373825073242,0.222147321701049,0.89799518585205,5.68776850700378]
cython_ = [0.0478317260742187,0.177917575836181,0.688255643844604,4.31203145980835]
threading_ = [0.5695775509,1.32467784881591,3.02220244407653,10.9004221439361]
odeint_ = [1.34515214,5.33712148666381,24.7575545310974,151.328786849975]
list_of_N = [100,200,400,1000]
line1, = plt.plot(list_of_N, verlet, 'g', linewidth=1, label='verlet')
line2, = plt.plot(list_of_N, odeint_, 'y', linewidth=1, label="verlet-odeint")
line3, = plt.plot(list_of_N, threading_, 'r', linewidth=1, label="verlet-threading")
line4, = plt.plot(list_of_N, cython_, 'b', linewidth=1, label="verlet-cython")
plt.xlabel('Кол-во частиц')
plt.ylabel('Время')
plt.legend(handler_map={line1: HandlerLine2D(numpoints=4)})
fig = plt.figure(figsize=(20, 6))
plt.savefig('results.pdf', bbox_inches='tight')
plt.show()
```
